```python
# Your answer here (python code + explanation):


# Your answer here (python code + explanation):

import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt 
import random
import pandas as pan
import matplotlib


df = pan.read_csv ('https://raw.githubusercontent.com/suthapalliuday/INFO5502-Summer2022/main/datasets/lab_assignment_05_census.csv')
for i in df.index:  # replacing 1 with male and 2 with female in our dataset
  if df['Sex'][i]==1:
     df['Sex'][i]='Male'
  else:
    df['Sex'][i]='Female'

sns.set(rc={'figure.figsize':(25,10)}) # settign the output size of the scatterplot
plt.ticklabel_format(style='plain', axis='y') # To print the complete number on y axis instead of e^7
ax = sns.scatterplot(x="Age", y="People", data=df, hue = 'Sex', size = 'Year') # this plots our graph
plt.ylabel("Number of People") # renaming Y lable
```

```python

year1900 =df[df['Year'] == 1900] # contains data of year 1900
year2000=df[df['Year'] == 2000]  # contains data of year 2000

fig, axes = plt.subplots(1, 2, sharex=True, figsize=(10,5)) # setting the axes to plot two graphs side by side
sns.barplot(ax=axes[0],x="Age", y="People", hue="Sex", data=year1900)
axes[0].set_title('1900')
sns.barplot(ax=axes[1],x="Age", y="People", hue="Sex", data=year2000)
axes[1].set_title('2000')

sns
```
